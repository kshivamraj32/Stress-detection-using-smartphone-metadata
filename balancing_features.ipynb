```python
import numpy as np
import pandas as pd
import os
from scipy import spatial

path ="G:/user_data/user_data/user"
open_path = ""
write_path = ""
data = []
user_id = []
for i in range(1,55):
    if i<10 :
        open_path = path + str(0) + str(i) + "/user" + str(i) + "_c.txt"
        write_path = path + str(0) + str(i) + "/user" + str(i) + "_cb.txt"
    else :
        open_path = path + str(i) + "/user" + str(i) + "_c.txt"
        write_path = path + str(i) + "/user" + str(i) + "_cb.txt"
    if os.path.exists(open_path) : 
        data1 = np.loadtxt(open_path,delimiter = " ")
        if data1.ndim == 1:
            data1.shape = (1,10)
        data1 = data1[:,0:9]
        data.append(np.mean(data1,axis = 0))
        user_id.append(i)
data = np.array(data)

# data1 = np.loadtxt("G:/user_data/user_data/user01/user1_c.txt",delimiter = " ");
# data1 = data1[:,0:9]
# data2 = np.loadtxt("G:/user_data/user_data/user02/user2_c.txt",delimiter = " ");
# data2 = data2[:,0:9]
# data = [np.mean(data1,axis = 0),np.mean(data2,axis = 0)]
# data = np.array(data)
```

```python
from math import sqrt
import operator
#print (sqrt(np.inner(data[0]-data[7] , data[0]-data[7]))  )
tr_path = "G:/user_data/user_data/user"
trr_path = ""
write_path = ""
for ind in range(len(user_id)):
    dists = []
    for ind2 in range(len(user_id)):
        if (ind2 != ind):
            dists.append(   (user_id[ind2],sqrt(np.inner(data[ind]-data[ind2] , data[ind]-data[ind2])) ) )
    dists.sort(key = operator.itemgetter(1))
    
    if user_id[ind] < 10 :
        trr_path = tr_path +  str(0) + str(user_id[ind]) + "/user" + str(user_id[ind]) + "_c.txt"
        write_path = tr_path + str(0) + str(user_id[ind]) + "/user" + str(user_id[ind]) + "_cb.txt"
    else : 
        trr_path = tr_path + str(user_id[ind]) + "/user" + str(user_id[ind]) + "_c.txt"
        write_path = tr_path + str(user_id[ind]) + "/user" + str(user_id[ind]) + "_cb.txt"
        
    f = open(write_path,"w+")
    rf = open(trr_path,"r")
    f.write(rf.read())
    rf.close()
    
    for x in range(4):
        if dists[x][0] < 10:
            trr_path = tr_path +  str(0) + str(dists[x][0]) + "/user" + str(dists[x][0]) + "_c.txt"
        else:
            trr_path = tr_path +  str(dists[x][0]) + "/user" + str(dists[x][0]) + "_c.txt"
            
        rf= open(trr_path,"r")
        f.write(rf.read())
        rf.close()
    f.close()
    print("Done ",user_id[ind])
    
            
            
        
        
   
```
